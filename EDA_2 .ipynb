```python
# Import necessary libraries
import pandas as pd
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# Identify the numerical columns
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns

# Standard Scaling
scaler_standard = StandardScaler()
df_standard_scaled = df.copy()
df_standard_scaled[numerical_cols] = scaler_standard.fit_transform(df[numerical_cols])

# Min-Max Scaling
scaler_minmax = MinMaxScaler()
df_minmax_scaled = df.copy()
df_minmax_scaled[numerical_cols] = scaler_minmax.fit_transform(df[numerical_cols])

# Display the first few rows of the scaled data
print("Standard Scaled Data:")
print(df_standard_scaled.head())

print("\nMin-Max Scaled Data:")
print(df_minmax_scaled.head())
```

Standard Scaling maintains the distribution shape and variance, making it preferable when the model is sensitive to the distribution and magnitude of differences between features.

Min-Max Scaling is better suited for non-Gaussian data or when you need to confine your data to a specific range, particularly when absolute differences matter (as in distance-based models).

Encoding Techniques

```python
# Import necessary libraries
import pandas as pd
from sklearn.preprocessing import OneHotEncoder, LabelEncoder

# Identify categorical columns
categorical_cols = df.select_dtypes(include=['object']).columns

# Loop through each categorical column and apply appropriate encoding
for col in categorical_cols:
    num_unique = df[col].nunique()  # Get the number of unique categories

    if num_unique <= 5:
        # Apply One-Hot Encoding if categories are 5 or less
        one_hot = pd.get_dummies(df[col], prefix=col)
        df = pd.concat([df, one_hot], axis=1)
        df.drop(col, axis=1, inplace=True)
    else:
        # Apply Label Encoding if categories are more than 5
        label_encoder = LabelEncoder()
        df[col] = label_encoder.fit_transform(df[col])

# Display the first few rows of the transformed data
print(df.head())
```

One-Hot Encoding pros and cons
pros
- No ordinal relationship imposed
- Widely accepted by many algorithms
- Prevents bias due to magnitude

cons
- High dimensionality
- Sparse representation
- Risk of overfitting with small datasets and many categories

Label Encoding Pros and cons
pros
- Efficient representation
- Ideal for ordinal data
- Simple to implement

cons
- Assumes ordinal relationship (can be misleading for nominal data)
- Bias in distance-based algorithms

```python
# Feature 1: Create 'work_experience_level'
# Assuming 'age' and 'education-num' columns are present in the dataset
df['work_experience_level'] = df['age'] - df['education_num']

# Feature 2: Create 'family_size'
# Assuming the 'relationship' column contains relationship information
# We'll classify based on relationships (Husband, Wife, etc.)

relationship_map = {
    0: 'Husband',
    1: 'Not-in-family',
    2: 'Own-child',
    3: 'Unmarried',
    4: 'Wife',
    5: 'Other-relative'
}
df['relationship'] = df['relationship'].map(relationship_map)

# Define a mapping for family sizes
family_size_mapping = {
    'Husband': 4,   # Assumption: Husband has a spouse and possibly two children
    'Wife': 4,      # Similar assumption for Wife
    'Own-child': 3, # Own-child might imply 2 parents, 1 child
    'Unmarried': 1, # Unmarried could imply 1 person
    'Other-relative': 2, # Assumption: 2 people in household
    'Not-in-family': 1    # Single individual
}

df['family_size'] = df['relationship'].map(family_size_mapping)
print(df[['relationship', 'family_size']].head())


# Optional: Check for missing values
print("Missing values in family_size:", df['family_size'].isnull().sum())
```

```python
# Check the distribution of the 'hours-per-week' feature
sns.histplot(df['hours_per_week'], bins=30)
plt.title('Distribution of Hours Per Week')
plt.xlabel('Hours Per Week')
plt.ylabel('Frequency')
plt.show()

# Apply log transformation to 'hours-per-week'
# Add a small constant (1) to avoid log(0) issues
df['log_hours_per_week'] = np.log1p(df['hours_per_week'])

# Check the distribution after transformation
sns.histplot(df['log_hours_per_week'], bins=30)
plt.title('Log Transformed Distribution of Hours Per Week')
plt.xlabel('Log of Hours Per Week')
plt.ylabel('Frequency')
plt.show()

# Display the first few rows to verify the new feature
print(df[['hours_per_week', 'log_hours_per_week']].head())
```

Justification for Log Transformation:
-Handling Skewness: The hours-per-week feature often exhibits a right-skewed distribution, meaning there are a few individuals who work a significantly higher number of hours compared to the majority. Log transformation reduces this skewness by compressing the range of higher values.

-Improving Model Performance: Many machine learning algorithms assume that input features are normally distributed. By applying a log transformation, we can help meet this assumption, potentially leading to better model performance and more reliable predictions.

-Interpretability: The log-transformed values can be interpreted as a percentage increase. For instance, a change in log hours can be interpreted as multiplicative changes in the original hours, which can provide insights into the impact of hours worked on income.

Feature Selection

```python
import pandas as pd
from sklearn.ensemble import IsolationForest

# Select relevant numerical features for outlier detection
# For this example, we'll use 'hours-per-week' and 'age'
features = df[['hours_per_week', 'age']]

# Initialize the Isolation Forest model
iso_forest = IsolationForest(contamination=0.05, random_state=42)  # Assuming 5% outliers

# Fit the model and predict outliers
df['outlier'] = iso_forest.fit_predict(features)

# Remove outliers (label -1 indicates an outlier)
df_cleaned = df[df['outlier'] != -1]

# Display the shape before and after removing outliers
print(f"Original dataset shape: {df.shape}")
print(f"Cleaned dataset shape: {df_cleaned.shape}")

# Optionally, remove the 'outlier' column from the cleaned dataset
df_cleaned = df_cleaned.drop(columns='outlier')

# Display the first few rows of the cleaned dataset
print(df_cleaned.head())

```
